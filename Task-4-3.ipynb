В каком количестве воды (в литрах) надо растворить 400 г. соли, чтобы получить 5% раствор соли?

```python
ms = 400
ps = 5
mr = (ms * 100) / ps
mh = mr - ms
print(mh)

```

```python
from sympy.abc import x
from sympy import solve
from sympy import Eq

ans = solve(Eq(400 / (400 + x) * 100, 5), x)
print(ans)
```
